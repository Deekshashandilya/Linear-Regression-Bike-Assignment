### Dropping Variables

- **Based on the above observation , working day and weekdays seems to tend to variance , so dropping both Workingday and Weekdays**
    - `dteday` is representing date and year , which are already specified in `yr` and `mnth` cloumns
    - `Casual` and `Registered` variables only define type of customer , but their sum already defined by `cnt`
    - `instant` is index variable , not required for training purposue

```python
drop_list = ['instant','dteday','workingday','casual','registered','atemp', 'weekday'] #not required (dteday and holiday)
```

```python
bike_data = bike_data.drop(drop_list , axis = 1)
```

```python
bike_data.head()
```

- categorical variables :: seasons , yr,mnth,weekday, workingday,weathersit 
    - these all are categorical data in form of numeric , so what we need to do is first prepare the data for dummy creation then , dummy variable creation

```python
def season_map(x):
    return x.map({1:'spring', 2:'summer', 3:'fall', 4:'winter'})

bike_data['season'] = bike_data[['season']].apply(season_map)
```

```python
bike_data['season'].value_counts()
```

```python
def month_map(x):
    return x.map({1:'Jan',2:'Feb',3:'March', 4:'April', 5:'May' , 6:'June', 7:'July',
                 8:'August', 9:'Sept' , 10:"October" , 11:"Nov" , 12:'Dec'})

bike_data['mnth'] = bike_data[['mnth']].apply(month_map)
```

```python
bike_data['mnth'].value_counts()
```

```python
def weather_map(x):
    return x.map({1:'Good_Weather',2:'Cloudy',3:'Light_Snow', 4:'Heavy_Rain'})

bike_data['weathersit'] = bike_data[['weathersit']].apply(weather_map)
```

```python
bike_data['weathersit'].value_counts()
```

```python
bike_data.head()
```

### Dummy Variable 

#### Dummy variables for season

```python
# Get the dummy variables for the feature 'furnishingstatus' and store it in a new variable - 'status'

df_season = pd.get_dummies(bike_data['season'], drop_first = True)

# Check what the dataset 'status' looks like
df_season.head()
```

#### Dummy variables for Month

```python

df_month = pd.get_dummies(bike_data['mnth'], drop_first = True)

# Check what the dataset 'status' looks like
df_month.head()
```

```python
# ?pd.get_dummies
```

#### Dummy variables for Weather

```python
df_weather = pd.get_dummies(bike_data['weathersit'], drop_first = True)

# Check what the dataset 'status' looks like
df_weather.head()
```

#### Concatinating the values

```python
bike_data
```

```python
bike_data = pd.concat([bike_data, df_season,df_weather, df_month], axis = 1)
```

```python
bike_data.head()
```

```python
bike_data.columns
```

#### Deleting Duplicate Columns

```python
# Drop 'furnishingstatus' as we have created the dummies for it
bike_data.drop(['season','mnth','weathersit'], axis = 1, inplace = True)

bike_data.head()
```

## Splitting the Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively

df_train, df_test = train_test_split(bike_data, train_size = 0.7, test_size = 0.3, random_state = 100)
```

## Rescaling the Features

```python
# from sklearn.preprocessing import MinMaxScaler
# scaler = MinMaxScaler()

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
numeric_vars = df_train.columns

df_train[numeric_vars] = scaler.fit_transform(df_train[numeric_vars])

df_train.head()
```

## Building our model

This time, we will be using the **LinearRegression function from SciKit Learn** for its compatibility with RFE (which is a utility from sklearn)

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 14)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

- Values with True RFE

```python
col_values = X_train.columns[rfe.support_]
col_values
```

- Values with False RFE

```python
neg_value_col = X_train.columns[~rfe.support_]
neg_value_col
```

### Building model using statsmodel, for the detailed statistics

### Buliding the Model - I

```python
# Creating X_test dataframe with RFE selected variables
X_train_1 = X_train[col_values]
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_1 = sm.add_constant(X_train_1)
```

```python
lm = sm.OLS(y_train,X_train_1).fit()   # Running the linear model
```

```python
#Let's see the summary of our linear model
lm.summary()
```

- R-squared value = 0.844
- Adj. R-square = 0.840
- F-statistics = 191
- Variables p-values
    - p-value of Nov: 0.144 and Jan: 0.054 are high

### Checking the VIF of variables
- Dropping the const variables

```python
X_train_1.columns
```

```python
X_train_new = X_train_1.drop(['const'], axis=1)
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# Let's check the correlation coefficients to see which variables are highly correlated

plt.figure(figsize = (16, 10))
sns.heatmap(X_train_new.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

## Residual Analysis of the train data

So, now to check if the error terms are also normally distributed (which is infact, one of the major assumptions of linear regression), let us plot the histogram of the error terms and see what it looks like.

```python
y_train_pred = lm.predict(X_train_lm)
```

```python
# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(y_train,y_train_pred)
fig.suptitle('y_train vs y_train_pred', fontsize=20)              # Plot heading 
plt.xlabel('y_train', fontsize=18)                          # X-label
plt.ylabel('y_train_pred', fontsize=16)                          # Y-label
```

```python
res = y_train - y_train_pred
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot(res, bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```

```python
from sklearn.metrics import r2_score
r2_score(y_train, y_train_pred)
```
